Construa um programa onde o usuário digitará o seu peso (Kg) e sua altura (cm) e será calculado o IMC.

```python
peso = float(input("Digite o seu peso (kg): "))
print(f"O seu peso é {peso} Kg.: ")

altura = float(input("Digite a sua altura (cm): "))
print(f"A sua altura é {altura} cm.")
```

Cálculo do IMC = peso / altura * altura

```python
imc = peso / (altura*altura)
print(f"O seu IMC é igual a {imc}")
```
