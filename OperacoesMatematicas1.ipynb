Soma, subtração e multiplicação

```python
n1 = 6
n2 = 3

soma = n1 + n2
print(soma)

subtracao = n1 - n2
print(subtracao)

multiplicacao = n1 * n2
print(multiplicacao)
```

Divisão

```python
n3 = 20
n4 = 3

divisao = n3 / n4 #divisão com ponto flutuante
print(divisao)

divisa = n3 // n4 # divisão exata com resto
print(divisa)

div = n3 % n4  # resto da divisão
print(div)
```
