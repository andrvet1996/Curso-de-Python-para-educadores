Construa um programa onde o usuário digitará o seu peso (Kg) e sua altura (cm) e será calculado o IMC. Classifique o indivíduo de acordo com seu IMC.

```python
peso = float(input("Digite o seu peso (kg): "))
print(f"O seu peso é {peso} Kg.: ")

altura = float(input("Digite a sua altura (cm): "))
print(f"A sua altura é {altura} cm.")
```

Cálculo do IMC = peso / altura**2

```python
imc = peso / (altura**2)
print(f"O seu IMC é igual a {imc:.2f}.")
```

Classificação do IMC em peso adequado, abaixo do peso e acima do peso.

image.png

```python
if imc < 18.50:
  print("Você está ABAIXO do peso adequado. Procure o seu médico!")
elif imc > 24.99:
  print("Você está ACIMA do peso adequado. Procure o seu médico!")
else:
  print("O seu peso está ADEQUADO. Fique tranquilo!")

```
