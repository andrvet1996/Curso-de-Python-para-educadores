O usuário digitará um número. Se esse número for maior do que 5 aparecerá a mensagem "Você digitou um número maior do que 5". Se for o número 5 aparecerá a mensagem "Você digitou o número 5". Se o número digitado for menor que 5 aparecerá a mensagem "Você digitou um número menor do que 5".

```python
num = int(input("Digite um número: "))
```

```python
if num > 5:
  print("Você digitou um número maior do que 5.")
elif num == 5:
  print("Você digitou o número 5.")
else:
  print("Você digitou um número menor do que 5.")
```
