# ** Calculadora de diluição (Ci . Vi = Cf . Vf)**

```python
Ci = float(input("Digite a concentração inicial: "))
```

```python
Vi = float(input("Digite o volume inicial: "))
```

```python
Vf = float(input("Digite o volume final: "))
```

```python
Cf = (Ci * Vi)/Vf
print(f'A concentraçao final será de {Cf} g/dL')
```
