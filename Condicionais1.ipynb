O usuário digitará um número. Se esse número for maior do que 5 aparecerá a mensagem "Você digitou um número maior do que 5". Se o número digitado for menor ou igual a 5 nada aparecerá.

```python
num = int(input("Digite um número: "))
```

```python
if num > 5:
  print("Você digitou um número maior do que 5.")
```
